```python
import matplotlib.pyplot as plt
x_train = [1, 3, 5 , 7, 8]
y_train = [3, 5 , 4, 6, 7]


plt.scatter(x_train, y_train)
plt.xlim(xmin=0)
plt.ylim(ymin=0)
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.title("Test Data")

plt.show()
```

```python
#y = a*x + b
def loss(a, b):
  loss = 0
  for i in range(len(x_train)):
    loss += (y_train[i] - (a * x_train[i] + b))**2
  return loss / len(x_train)

```

```python
#deriative
#y = -2 * (y - (a*x + b))
#y = -2 * b * (y - (a*x + b))
from random import randint
import numpy as np


steps = 0
lr = 0.01
a = 2
b = 1

prev_loss = float('inf')

while steps < 10000:
  steps += 1
  random = randint(0, len(x_train) - 1)

  steps_b =  -2 * (y_train[random] - (a * x_train[random] + b))
  steps_a = -2 * x_train[random] * (y_train[random] - (a * x_train[random] + b))

  a -= lr * steps_a
  b -= lr * steps_b

  current_loss = loss(a, b)

  if abs(prev_loss - current_loss) < 1e-4:
    break

  prev_loss = current_loss

print(a, b)

print(a, b)
x_line = np.array(x_train)

# Calculate y values for the line using the final intercept and slope (0.6)
y_line = a * x_line + b
# Plot the original scatter plot
plt.scatter(x_train, y_train, label='Original Data')

# Plot the fitted line
plt.plot(x_line, y_line, color='red', label='Fitted Line')

plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.xlim(xmin=0)
plt.ylim(ymin=0)
plt.title("Test Data with Fitted Line")
plt.legend()
plt.show()
```
