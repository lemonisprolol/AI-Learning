```python
import matplotlib.pyplot as plt
weight  = [1, 2, 4, 3 , 6,  5, 8 , 12, 9, 11]
obesity = [0, 0 ,0, 1 , 0 , 1, 1 , 1 , 1, 1]
n = len(weight)
plt.scatter(weight, obesity)
plt.xlim(xmin=0)
plt.show()
```

```python
steps = 0
a = -2
b = 5
def sigmoid(z):
  return 1 / (1 + math.exp(-z))

lr = 0.03
while steps < 100000:
  steps += 1
  predictions = []
  for i in range(0, n):
    predictions.append(sigmoid(weight[i] * a + b))

  da = 0
  db = 0
  for i in range(0, n):
    error = predictions[i] - obesity[i]
    da += error * weight[i]
    db += error

  da /= n
  db /= n

  a -= lr * da
  b -= lr * db

print(f"Optimal 'a': {a}")
print(f"Optimal 'b': {b}")
```

```python
import numpy as np
plt.scatter(weight, obesity)
plt.xlim(xmin=0)

x_values = np.linspace(min(weight) - 1, max(weight) + 1, 100)
y_values = [sigmoid(w * a + b) for w in x_values]

plt.plot(x_values, y_values, color='red', label='Logistic Regression Curve')
plt.xlabel("Weight")
plt.ylabel("Probability of Obesity")
plt.title("Logistic Regression Fit")
plt.legend()
plt.grid(True)
plt.show()
```
