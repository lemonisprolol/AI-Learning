```python
import matplotlib.pyplot as plt
import numpy as np

x_train = [0, 5]
y_train = [1, 2]

plt.scatter(x_train, y_train, marker='o')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.grid(True)

plt.xlim(xmin=0)
plt.ylim(ymin = 0)
y1 = 0
for i in y_train:
  y1 += i
y1 /= len(y_train)
# Add horizontal line
x_line = np.array([0, max(x_train)])  # Line spanning the x-axis range
y_line = np.array([y1, y1])        # Constant y-value of 2.5
plt.plot(x_line, y_line, color='red', linestyle='-', label=f'y = {y1}')

plt.legend() # Show legend
plt.show()
```

```python
def loss(x, y, a, b):
  loss = 0
  for i in range(len(x)):
    loss += (y[i] - (x[i] * a + b))**2
  return loss/len(x)
# (var1 - var2) / var1
var1 = loss(x_train, y_train, 0, y1)
var2 = (loss(x_train, y_train,0.20001206411035122 , 0.9999414109448249))
print((var1 - var2)/ var1 * 100)
```
